## Count Airports per Country

### Subtask:
Count the number of airports for each country in the dataframe.

````python
filtered_airport_counts = airport_counts[airport_counts > 20]
print(filtered_airport_counts.head())
````

## Filter Countries

### Subtask:
Filter the countries to include only those with more than 15 airports.


**Reasoning**:
To filter the countries with more than 15 airports, I will apply a boolean mask to the `airport_counts` Series, selecting only values greater than 15. The result will be stored in `filtered_airport_counts`.



````python
filtered_airport_counts = airport_counts[airport_counts > 15]
print(filtered_airport_counts.head())
````

## Filter DataFrame

### Subtask:
Filter the original dataframe to keep only the entries corresponding to countries with more than 15 airports.


**Reasoning**:
To filter the original DataFrame, I will first extract the country names (index) from the `filtered_airport_counts` Series, which contains countries with more than 15 airports. Then, I will use these country names to filter the `df` DataFrame.



````python
filtered_countries_list = filtered_airport_counts.index.tolist()
df_filtered_countries = df[df['Country'].isin(filtered_countries_list)]
print(df_filtered_countries.head())
````

## Generate Bar Chart for Filtered Data

### Subtask:
Generate a bar chart showing the number of airports in countries that have more than 15 airports. Ensure the chart has appropriate labels and a legend.


**Reasoning**:
To visualize the number of airports in countries with more than 15 airports, I will use seaborn and matplotlib to create a bar chart from the `filtered_airport_counts` Series. I will set the title, x-axis label, y-axis label, and rotate the x-axis ticks for readability.



````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 7))
sns.barplot(x=filtered_airport_counts.index, y=filtered_airport_counts.values)
plt.title('Number of Airports in Countries with More Than 15 Airports')
plt.xlabel('Country')
plt.ylabel('Number of Airports')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````
